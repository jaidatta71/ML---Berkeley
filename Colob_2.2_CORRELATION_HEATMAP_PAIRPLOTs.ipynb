```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
cars = pd.read_csv('/content/drive/MyDrive/cars.csv')
cars.info()
```

```python
num_cars = cars.select_dtypes(include= 'number')
num_cars.info()
num_cars.head(10)
```

```python
#sns.pairplot(num_cars, hue = 'vs', palette = "Set2")
#plt.savefig('/content/drive/MyDrive/Images/carspaiplot.png')

g = sns.pairplot(num_cars, hue='vs', palette='Set1', diag_kind='kde', plot_kws={'s': 200})
g.fig.suptitle('Pairplot for v-shaped and straight Engine', y=1.02, fontsize=30)
plt.show()

g = sns.pairplot(num_cars, hue='cyl', palette='Set1', diag_kind='kde', plot_kws={'s': 200})
g.fig.suptitle('Pairplot for Cylinders', y=1.02, fontsize=30)
plt.show()


g = sns.pairplot(num_cars, hue='am', palette='Set1', diag_kind='kde', plot_kws={'s': 200})
g.fig.suptitle('Pairplot for Auto/Manual', y=1.02, fontsize=30)
plt.show()

plt.savefig('/content/drive/MyDrive/Images/carspaiplot.png')
```

```python
sns.heatmap(num_cars.corr(), annot=True)
plt.savefig('/content/drive/MyDrive/Images/cars.png')
```
