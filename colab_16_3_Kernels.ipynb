### Problem 1

#### Logistic Regression with Linear Kernel



Below, you are provided two functions -- `Kernel_matrix` and `evaluate_kernel_model` -- that are taken from the lectures.  Use these functions together with the function `linear_kernel` to determine the following variables:


- `linear_kernel_matrix`: Defined using the `Kernel_matrix` function with arguments `linear_kernel_function` and `X_train`.
- `linear_logistic`: Fit a `LogisticRegression` model (using `max_iter=1000`) on the `linear_kernel_matrix` and `y_train`
- `linear_predictions`: Use the `evaluate_kernel_model` function with the model, `X_train` and `X_test`




```python
def Kernel_matrix(kfunc, X):
    N, _ = X.shape
    K = np.empty((N, N))
    for i in range(N):
        for j in range(N):
            K[i, j] = kfunc(X[i, :], X[j, :])
    return K
```

## Purpose according to Gemini:

This function calculates the Kernel matrix, which is a fundamental component in kernel-based machine learning methods like Support Vector Machines (SVMs). The Kernel matrix represents the similarity between all pairs of data points in a dataset using a kernel function.

Step-by-step explanation:

def Kernel_matrix(kfunc, X):: This line defines the function named Kernel_matrix that accepts two arguments:

kfunc: This is the kernel function that will be used to compute the similarity between data points. It could be a linear kernel, polynomial kernel, radial basis function (RBF) kernel, or any other custom kernel.
X: This is the input data matrix where each row represents a data point and each column represents a feature.
N, _ = X.shape: This line gets the number of data points (N) in the input data matrix X. The underscore _ is used to ignore the number of features, as it's not needed in this function.

K = np.empty((N, N)): This line creates an empty NumPy array K with dimensions N x N. This array will store the Kernel matrix.

for i in range(N):: This loop iterates through each data point in the dataset.

for j in range(N):: This nested loop iterates through each data point again, allowing us to compare all possible pairs of data points.

K[i, j] = kfunc(X[i, :], X[j, :]): This is the core of the function. It calculates the kernel value between the i-th and j-th data points using the provided kernel function kfunc. The result is stored in the K matrix at the corresponding position (i, j).

return K: Finally, the function returns the computed Kernel matrix K.

In essence, the Kernel_matrix function applies the given kernel function to all possible pairs of data points in the input data X and stores the results in a matrix, which is then returned. This Kernel matrix can then be used by other machine learning algorithms for tasks like classification or regression.

```python
def evaluate_kernel_model(model, kfunc, X_train, X_test):
    N1, _ = X_train.shape
    N2, _ = X_test.shape

    K = np.empty((N2, N1))
    for i in range(N2):
        for j in range(N1):
            K[i, j] = kfunc(X_train[j, :], X_test[i, :])
    return model.predict(K)
```

## Purpose according to Gemini:

This function is used to evaluate a kernel-based model on a test dataset. It essentially predicts the target values for the test data using the trained model and the kernel function.

Step-by-step explanation:

def evaluate_kernel_model(model, kfunc, X_train, X_test):: This line defines the function named evaluate_kernel_model that accepts four arguments:

model: The trained kernel-based model (e.g., a Support Vector Machine) that will be used for prediction.
kfunc: The kernel function that was used to train the model.
X_train: The training data used to train the model.
X_test: The test data for which predictions are to be made.
N1, _ = X_train.shape: This line gets the number of training data points (N1).

N2, _ = X_test.shape: This line gets the number of test data points (N2).

K = np.empty((N2, N1)): This line creates an empty NumPy array K with dimensions N2 x N1. This array will store the kernel matrix between the test data and the training data.

for i in range(N2):: This loop iterates through each test data point.

for j in range(N1):: This nested loop iterates through each training data point.

K[i, j] = kfunc(X_train[j, :], X_test[i, :]): This is the core of the function. It calculates the kernel value between the j-th training data point and the i-th test data point using the provided kernel function kfunc. This value represents the similarity between these two data points and is stored in the K matrix.

return model.predict(K): Finally, the function uses the trained model's predict method to predict the target values for the test data based on the calculated kernel matrix K. These predictions are then returned by the function.

In essence, the evaluate_kernel_model function calculates the similarity between the test data and the training data using the kernel function and then uses this similarity information along with the trained model to make predictions for the test data. This allows us to evaluate the performance of the kernel-based model on unseen data.

```python
def linear_kernel_function(x, z):
    return np.dot(x, z)
```

## Purpose according to Gemini:

This function calculates the linear kernel between two data points, represented by vectors x and z. The linear kernel is a simple measure of similarity between two vectors, calculated as their dot product.

Step-by-step explanation:

def linear_kernel_function(x, z):: This line defines the function named linear_kernel_function that accepts two arguments:

x: The first data point, represented as a vector.
z: The second data point, represented as a vector.
return np.dot(x, z): This line calculates the dot product of the two input vectors x and z using the np.dot function from the NumPy library. The dot product is a measure of similarity between two vectors, calculated as the sum of the products of their corresponding elements. The result of the dot product is then returned by the function.

In essence, the linear_kernel_function calculates the linear kernel between two data points by computing their dot product. This kernel function is often used in kernel-based machine learning algorithms like Support Vector Machines (SVMs) to measure the similarity between data points in a linear fashion.

```python

linear_kernel_matrix = Kernel_matrix(linear_kernel_function, X_train)
linear_logistic = LogisticRegression(max_iter = 1000).fit(linear_kernel_matrix, y_train)
linear_predictions = evaluate_kernel_model(linear_logistic, linear_kernel_function, X_train, X_test)



### ANSWER CHECK
print(linear_predictions[:10])
```

```python
## Uncomment to Visualize
xx = np.linspace(X_train[:, 0].min(), X_train[:, 0].max(), 50)
yy = np.linspace(X_train[:, 1].min(), X_train[:, 1].max(), 50)
XX, YY = np.meshgrid(xx, yy)

grid = np.c_[XX.ravel(), YY.ravel()]
labels = evaluate_kernel_model(linear_logistic, linear_kernel_function, X_train, grid)

plt.contourf(xx, yy, labels.reshape(XX.shape), cmap = 'tab10', alpha = 0.3)
sns.scatterplot(data = X, x = 'total_phenols', y = 'color_intensity', hue = y)
plt.xlim(1, 3.5)
plt.ylim(2, 10)
```

### Problem 2

#### Logistic Regression with Quadratic Kernel



Now, complete the `quadratic_kernel_function` below to implement a polynomial kernel of degree 2 on the input data similar to that of the `linear_kernel_function`.  

Then, use the `Kernel_matrix` function with arguments `quadratic_kernel_function` and `X_train` to generate a `quadratic_kernel_matrix` object.

Fit the `quadratic_kernel_matrix` and `y_train` to a logistic model as `quadratic_logistic` below -- to avoid a convergence warning set `max_iter = 1000`.  

Use the `evaluate_kernel_model` function with the appropriate arguments to calculate the predictions as `quadratic_predictions`.

Uncomment the code below to visualize the resulting decision boundary.

```python

def quadratic_kernel_function(x, z):
    '''
    This function applies a quadratic kernel to
    array x and z.
    '''
    return (np.dot(x,z) + 1)**2

quadratic_kernel_matrix = Kernel_matrix(quadratic_kernel_function, X_train)
quadratic_logistic = LogisticRegression(max_iter = 1000).fit(quadratic_kernel_matrix, y_train)
quadratic_predictions = evaluate_kernel_model(quadratic_logistic, quadratic_kernel_function, X_train, X_test)


### ANSWER CHECK
print(quadratic_predictions[:10])
```

```python
labels = evaluate_kernel_model(quadratic_logistic, quadratic_kernel_function, X_train, grid)

plt.contourf(xx, yy, labels.reshape(XX.shape), cmap = 'tab10', alpha = 0.3)
sns.scatterplot(data = X, x = 'total_phenols', y = 'color_intensity', hue = y)
plt.xlim(1, 3.5)
plt.ylim(2, 10)
plt.title('Quadratic Decision Boundary with Kernel');
```

### Problem 3

#### Logistic Regression with Quintic Kernel



Finally, complete the function `quintic_kernel_function` below that creates a polynomial kernel of degree 5.  Use this to create the `quintic_kernel_matrix`, `quintic_logistic`, and `quintic_predictions` in a similar way as in the previous questions.  

**NOTE**: To avoid convergence warnings here, set the `max_iter = 10_000` in your `quintic_logistic` estimator.

Uncomment the code to visualize the new decision boundary.

```python

def quintic_kernel_function(x, z):
    '''
    This function applies a quintic (5th degree
    polynomial) kernel to
    array x and z.
    '''
    return (np.dot(x,z) + 1)**5

quintic_kernel_matrix = Kernel_matrix(quintic_kernel_function, X_train)
quintic_logistic =  LogisticRegression(max_iter = 10_000).fit(quintic_kernel_matrix, y_train)
quintic_predictions = evaluate_kernel_model(quintic_logistic, quintic_kernel_function, X_train, X_test)



### ANSWER CHECK
print(quintic_predictions[:10])
```

```python
labels = evaluate_kernel_model(quintic_logistic, quintic_kernel_function, X_train, grid)
plt.contourf(xx, yy, labels.reshape(XX.shape), cmap = 'tab10', alpha = 0.3)
sns.scatterplot(data = X, x = 'total_phenols', y = 'color_intensity', hue = y)
plt.xlim(1, 3.5)
plt.ylim(2, 10)
plt.title('Decision Boundary with Polynomial Degree 5 (Quintic) Kernel');
```
