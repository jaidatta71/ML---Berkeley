### Problem 7: Scatterplot of GDP vs. Life Expectancy



`matplotlib` also has a scatterplot method that you can use with `plt.scatter()`.  Use this to create a scatterplot with the GDP column on the x-axis and Life Expectancy on the y-axis.  Add human-readable labels "GDP Per Capita" and "Life Expectancy (years)" respectively and a title "GDP vs. Life Expectancy".



````python
plt.scatter(x = gap_df['lifeExp'], y = gap_df['gdpPercap'], alpha = 0.6, color = 'skyblue', edgecolor = 'black', label = 'Life Expectancy')

#plt.hist(sample_means, bins=10, color='skyblue', edgecolor='black', label = 'Sample Size')
plt.title('GDP vs. Life Expectancy')
plt.ylabel('GDP per capita')
plt.xlabel('Life Expectancy (years)')
plt.grid()
plt.show()
plt.savefig('/content/drive/MyDrive/colab_activity2_3_starter/images/gdp_scatter.png')
````

### Problem 8: Logarithmic Transformation



Notice the non-linear relationship between GDP and Life Expectancy.  One common transformation is to take something that looks exponential and take a logarithm in order to transform the relationship into a linear one.  Draw a scatterplot like the above, but use `np.log()` to plot the logarithm of GDP vs. Life Expectancy.  Add appropriate x and y labels, and the title "Logarithm of GDP vs. Life Expectancy".



````python
plt.scatter(x = np.log(gap_df['lifeExp']), y = np.log(gap_df['gdpPercap']), alpha = 0.6, color = 'skyblue', edgecolor = 'black', label = 'Life Expectancy')

#plt.hist(sample_means, bins=10, color='skyblue', edgecolor='black', label = 'Sample Size')
plt.title('GDP vs. Life Expectancy')
plt.ylabel('GDP per capita')
plt.xlabel('Life Expectancy (years)')
plt.grid()
plt.show()
plt.savefig('/content/drive/MyDrive/colab_activity2_3_starter/images/gdp_scatter.png')
````

### Problem 10: `pandas` scatterplot



Similar to the `.hist()` method, pandas has a general `.plot()` method where you can specify the kind of plot.  Use this to create a scatterplot of GDP on the x-axis and Life Expectancy on the y-axis.  You will have to call `.plot()` on the entire DataFrame and specify the `x` and `y` arguments to the `.plot()` function with the `kind = scatter` argument.  Save your plot as `gdp_lifexp_scatter_pandas.png` to the `images` folder.



````python

gap_df['gdpPercap'].hist(bins = 10, alpha = 0.6, color = 'lightblue', edgecolor = 'black', label = 'Life Expectancy', grid = False)
gap_df.hist(column = 'gdpPercap', bins = 10, alpha = 0.6, color = 'lightblue', edgecolor = 'black', label = 'Life Expectancy', grid = False)

plt.scatter(x = np.log(gap_df['lifeExp']), y = np.log(gap_df['gdpPercap']), alpha = 0.6, color = 'skyblue', edgecolor = 'black', label = 'Life Expectancy')
gap_df.plot(kind = 'scatter', x = 'gdpPercap', y = 'lifeExp', alpha = 0.6, color = 'skyblue', edgecolor = 'black', label = 'Life Expectancy', grid = True)


#plt.hist(sample_means, bins=10, color='skyblue', edgecolor='black', label = 'Sample Size')

#plt.title('GDP vs. Life Expectancy')
#plt.ylabel('GDP per capita')
#plt.xlabel('Life Expectancy (years)')
#plt.grid()
#plt.show()
plt.savefig('/content/drive/MyDrive/colab_activity2_3_starter/images/gdp_lifexp_scatter_pandas.png')

````
