```python
# find outliers

def find_boundaries(df, variable, distance):
  IQR = df[variable].quantile(0.75) - df[variable].quantile(0.25)

  lower_boundary = df[variable].quantile(0.25) - distance*IQR
  upper_boundary = df[variable].quantile(0.75) + distance*IQR

  return lower_boundary, upper_boundary
```

```python
# Tukey formula = 1.5 x IQR
lower, upper = find_boundaries(baseball, "Salary", 1.5)
```

```python
import statsmodels.api as sm
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
import warnings
warnings.filterwarnings("ignore")
```

```python
y = data['Inbound_Calls']
exog = data[['Marketing_Campaign', "Holiday"]]
```

```python
X_train, X_test, y_train, y_test = train_test_split(exog, y, test_size = 0.2, shuffle=False)
```

```python
import itertools
# setup the order params
p = q = range(0,4)
d = [0,1,2]
s = 7 # weekly

pdq = list(itertools.product(p,d,q))
```

```python
pdq
```

```python
import numpy as np

best_mae = np.inf
best_order = None
best_seasonal_order = None
best_model = None
```

```python
for (p_,d_,q_) in pdq:
  order = (p_,d_,q_)
  seasonal_order = (p_,d_,q_,s)
  print(f"Running {seasonal_order}")
  try:
    # train the model
    model = sm.tsa.statespace.SARIMAX(
        y_train,
        exog = X_train,
        order = order,
        seasonal_order = seasonal_order,
        enforce_stationarity=False
    )
    results = model.fit(disp = False)

    forecast = results.get_forecast(steps=len(y_test), exog = X_test)
    y_pred = forecast.predicted_mean
    # compute the MAE
    mae = mean_absolute_error(y_test, y_pred)

    if mae < best_mae:
      best_mae = mae
      best_order = order
      best_seasonal_order = seasonal_order
      best_model = results
      print(f"New best model SARIMAX: {best_seasonal_order} -- MAE: {mae}")

  except:
    continue

```

```python
best_model.summary()
```

```python
# Inbound Calls = m1 * Marketing + m2 * Holiday + c + Seasonality + MA + AR
```

```python
data["Date"] = pd.to_datetime(data["Date"])
data = data.set_index("Date")
```

```python
forecast_length = 50
future_dates = pd.date_range("2025-01-01", periods=forecast_length, freq="D")
future_exog = pd.DataFrame({
    "Marketing_Campaign": np.random.choice([0,1], size=forecast_length, p=[0.8,0.2]),
    "Holiday": [1 if d.weekday() == 6 else 0 for d in future_dates]
}, index=future_dates)

# --- 2. Forecast ---
forecast_res = best_model.get_forecast(steps=forecast_length, exog=future_exog)
forecast = forecast_res.predicted_mean
conf_int = forecast_res.conf_int()

# --- 3. Plot ---
plt.figure(figsize=(12,5))
plt.plot(data.index, data["Inbound_Calls"], label="Observed")
plt.plot(future_dates, forecast, label="Forecast", color="red")
plt.fill_between(future_dates, conf_int.iloc[:,0], conf_int.iloc[:,1],
                 color="pink", alpha=0.3)
plt.title("Forecast of Inbound Calls with Exogenous Variables")
plt.legend()
plt.show()
```
